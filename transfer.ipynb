```python
import pandas, cv2, json, os
```

```python
annotation = pandas.read_csv("/home/dsd/Desktop/calendar_faster_rcnn/dataset/calendar.csv")
annotation
```

```python
path = '/home/dsd/Desktop/calendar_faster_rcnn/dataset/train/'
filename = annotation['filename']
coordinate = annotation['region_shape_attributes']
region_id = annotation['region_attributes']
# print(coordinate)
combine = zip(coordinate, region_id)

data = ''
for i, c in enumerate(combine):
    # print(i, c[0], c[1])
    full_filename = '/content/drive/calendar_faster_rcnn/dataset/train/' + filename[i]
    m, n = json.loads(c[0]), json.loads(c[1])['type']
    # print(m, n)
    x1, y1, x2, y2 = m['x'], m['y'], m['x'] + m['width'], m['y'] + m['height']
    data += full_filename + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + str(n) + '\n'

with open(path + 'annotation.txt', 'w') as f:
    f.write(data)
```
